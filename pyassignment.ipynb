```python
# prompt: Write a Python program that takes an input number from the user, converts it to different numeric data types (integer, float, and complex), and displays the converted values.
# Explain the difference between these data types in comments.

# Get input number from the user
num_str = input("Enter a number: ")

# Convert to integer
num_int = int(num_str)  # Integer: Represents whole numbers without decimals
print("Integer:", num_int)

# Convert to float
num_float = float(num_str)  # Float: Represents numbers with decimals
print("Float:", num_float)

# Convert to complex
num_complex = complex(num_str)  # Complex: Represents numbers with a real and imaginary part
print("Complex:", num_complex)
```

```python
# Prompt the user to enter the length and width of the rectangle.
# Calculate the area.

# Get input for length and width
w = float(input("Enter width of the rectangle: "))
l = float(input("Enter length of the rectangle: "))

# Calculate the area
area = l * w

# Display the result
print("The area of the rectangle is:", area)


#showing area in 2 decimal places
print(f"{area:.2f}")
```

```python
while True:
  user_input = input("Enter a number or 'exit': ")
  if user_input == 'exit':
    break
  else:
    try:
      number = float(user_input)
      if number > 0:
        print("Positive number")
      elif number < 0:
        print("Negative number")
      else:
        print("Zero")
    except ValueError:
      print("Invalid input. Please enter a number or 'exit'.")
    continue
```

```python
# Get input for two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Check if both numbers are even
if num1 % 2 == 0 and num2 % 2 == 0:
  print("Both numbers are even.")
# Check if both numbers are odd
elif num1 % 2 != 0 and num2 % 2 != 0:
  print("Both numbers are odd.")
# Otherwise, one is even and one is odd
else:
  print("One number is even, and the other is odd.")
```

```python
def convert_to_bases(number):
    bases = {
        'binary': '',
        'octal': '',
        'hexadecimal': ''
    }

    for i in range(31, -1, -1):
        bit = (number >> i) & 1
        bases['binary'] = bases['binary'] + str(bit)

        if i % 3 == 0:
            octal_digit = (number >> i) & 0b111
            bases['octal'] = bases['octal'] + str(octal_digit)

        if i % 4 == 0:
            hex_digit = (number >> i) & 0xf
            bases['hexadecimal'] = bases['hexadecimal'] + hex(hex_digit)[2:]

    # Remove leading zeros
    bases['binary'] = bases['binary'].lstrip('0') or '0'
    bases['octal'] = bases['octal'].lstrip('0') or '0'
    bases['hexadecimal'] = bases['hexadecimal'].lstrip('0') or '0'

    return bases

# Get input from user
number = int(input("Enter an integer: "))

# Convert the number to different bases
results = convert_to_bases(number)

# Print
print(f"Binary: {results['binary']}")
print(f"Octal: {results['octal']}")
print(f"Hexadecimal: {results['hexadecimal']}")
```
